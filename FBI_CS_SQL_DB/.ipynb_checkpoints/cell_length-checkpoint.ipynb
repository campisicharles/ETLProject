```python
#
#  notebook to return the cell lengths
#

# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
import time
import json

%matplotlib inline
import matplotlib.pyplot as plt

import os
import datetime
```

```python
#import csv files

edu_csv = "D:\Homework\ETLProject\FBI_CS_SQL_DB\Data\Edu.csv"

edu_df = pd.read_csv(edu_csv, encoding="utf-8")
edu_df.head()
```

```python
edu_rounded = edu_df.round(4)
edu_rounded.head()
```

```python
# Export file as a CSV, with the Pandas index and with the header....
edu_rounded.to_csv("D:\Homework\ETLProject\FBI_CS_SQL_DB\Data\edu_rounded.csv", index=True, header=True)
```

```python

# return the max length of characters in each column

measurer = np.vectorize(len)

res1 = measurer(edu_df.values.astype(str)).max(axis=0)
res1
```
