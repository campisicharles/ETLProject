```python
census_pd["Unemployment Rate"] = 100 * \
    census_pd["Unemployment Count"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["No Schooling Rate"] = 100 * \
    census_pd["People over 25 with No Schooling"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["High School Education Rate"] = 100 * \
    census_pd["High School Diploma"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["GED Rate"] = 100 * \
    census_pd["GED"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Associates Degree Rate"] = 100 * \
    census_pd["Associates Degree"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Bachelor's Degree Rate"] = 100 * \
    census_pd["Bachelor's Degree"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Master's Degree Rate"] = 100 * \
    census_pd["Master's Degree"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Professional Degree Rate"] = 100 * \
    census_pd["Professional Degree"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Doctoral Degree Rate"] = 100 * \
    census_pd["Doctoral Degree"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["English Only Speakers Rate"] = 100 * \
    census_pd["English Only Speakers"].astype(
        int) / census_pd["Population"].astype(int)

census_pd["Poverty Rate"] = 100 * \
    census_pd["People Living in Poverty"].astype(
        int) / census_pd["Population"].astype(int)


census_pd
```

```python
years = ["2012", "2013", "2014", "2015", "2016", "2017"]
    
# set up lists to hold reponse info
off_year = []
off_type = []
off_ct = []

base_url = "https://api.usa.gov/crime/fbi/sapi/api/data/arrest/states/offense/OH/all/"
year = 0
slash = "/"
post_url = "?API_KEY=qjxuYoCo1z10Fvy60mTLHOcftyK9sxevJw14QbvI"

# Loop through the list of years and perform a request for data on each
for year in years:
    response = requests.get(base_url + year + slash + year + post_url).json()['results'][0]
    
    for offense in fbi_offense:
           
        off_year.append(response['data_year'])
        off_type.append(offense)
        off_ct.append(response[offense])
```

```python
fbi_dict = {
    "Year": off_year,
    "Offense_Type": off_type,
    "Offense_Count": off_ct
}

fbi_data = pd.DataFrame(fbi_dict)
fbi_data.head()
```
